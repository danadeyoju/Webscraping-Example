```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://www.investing.com/commodities/silver-historical-data'
```

```python
headers={'User-Agent':'Mozilla/5.0'} #allows you to bypass the Error 403:Forbidden
```

```python
response=requests.get(url,headers=headers)
```

```python
response.status_code #If you get 200 it means your site has successfully downloaded
```

```python
response.content
```

```python
soup = BeautifulSoup(response.content,'html.parser')
```

```python
silver_table = soup.find_all('table',class_='genTbl closedTbl historicalTbl')
```

```python
len(silver_table) #ifd it says 1 it means you have the table you were looking for
```

```python
type(silver_table)
```

```python
silver_table=silver_table[0] #using[0] extracts the table element of the result set
```

```python
silver_table
```

```python
type(silver_table)
```

```python
for row in silver_table.find_all('tr'):
    for cell in row.find_all('td'):
        print(cell.text)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame()

i=0
for row in silver_table.find_all('tr'):
    j=0
    for cell in row.find_all('td'):
        df.loc[i,j]=cell.text
        j=j+1   
    i=i+1       
        
        
```
